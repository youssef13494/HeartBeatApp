```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Prepare features (X) and target (y)
X3 = concatenated_data.drop(187, axis=1)  # Replace '187' with the column index or name to remove target column
y3 = concatenated_data[187]  # Replace '187' with the column index or name for the target (Normal or Abnormal)

# Step 2: Split the data into training and testing sets
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=42)

# Step 3: Train the Random Forest model
model3 = RandomForestClassifier(random_state=42)  # You can adjust parameters like n_estimators, max_depth, etc.
model3.fit(X_train3, y_train3)

# Step 4: Make predictions
y_pred3 = model3.predict(X_test3)

# Step 5: Evaluate the model
accuracy3 = accuracy_score(y_test3, y_pred3)
print(f'Accuracy For Binary Classfication Model: {accuracy3 * 100:.2f}%')

print('Classification Report For Binary Classfication Model:')
print(classification_report(y_test3, y_pred3))

print('Confusion Matrix For Binary Calssification :')
print(confusion_matrix(y_test3, y_pred3))
```

**Work For Binary Classification Model in train**

```python
# Check if the last column's value is 0 or not
train_data3['Normal_or_Abnormal'] = train_data3.iloc[:, -1].apply(lambda x: 0 if x == 0 else 1)
```

```python
#Drop Column 187
train_data3.drop(columns=187, inplace=True)
```

```python
train_data3.head()
```

```python
#Count for Each class in column Normal_or_Abnormal
class_counts = train_data3['Normal_or_Abnormal'].value_counts()

# Display the counts
class_counts
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Prepare features (X) and target (y)
X6 = train_data3.drop('Normal_or_Abnormal', axis=1)  # Replace with the appropriate column name
y6 = train_data3['Normal_or_Abnormal']

# Step 2: Split the data into training and testing sets
X_train6, X_test6, y_train6, y_test6 = train_test_split(X6, y6, test_size=0.2, random_state=42)

# Step 2: Apply SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)  # Initialize SMOTE
X_train6, y_train6 = smote.fit_resample(X_train6, y_train6)  # Resample the data

# Step 3: Calculate the scale_pos_weight
# Number of samples in class 1 / Number of samples in class 0
class_0_count = (y_train6 == 0).sum()
class_1_count = (y_train6 == 1).sum()
ratio = class_0_count / class_1_count

# Step 4: Train the XGBoost model with scale_pos_weight
model6 = xgb.XGBClassifier(
    random_state=42, 
    use_label_encoder=False, 
    eval_metric='mlogloss', 
    scale_pos_weight=ratio  # Adjusting the weight of classes
)
model6.fit(X_train6, y_train6)

# Step 5: Make predictions
y_pred6 = model6.predict(X_test6)

# Step 6: Evaluate the model
accuracy6 = accuracy_score(y_test6, y_pred6)
print(f'Accuracy For Binary Classification Model using XGBoost: {accuracy6 * 100:.2f}%')

print('Classification Report:')
print(classification_report(y_test6, y_pred6))

print('Confusion Matrix:')
print(confusion_matrix(y_test6, y_pred6))
```
